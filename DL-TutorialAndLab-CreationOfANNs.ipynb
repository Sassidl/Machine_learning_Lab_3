# Create neural network with no hidden layer

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation="sigmoid")
]
)
```

```python
model.compile(
    optimizer = "adam",
    loss = "sparse_categorical_crossentropy",
    metrics = ["accuracy"]
)
```

```python
model.fit(X_train_flatten, y_train, epochs=5)
```

```python
X_test_flatten = X_test.reshape(len(X_test), 28*28)
```

# Create a neural network with one hidden layer 

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation="relu"),
    keras.layers.Dense(10, activation="sigmoid")
])
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(X_train_flatten, y_train, epochs=5)
```

```python
model.evaluate(X_test_flatten, y_test)
```

```python
y_predicted = model.predict(X_test_flatten)
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
conf_matrix = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
```

```python
conf_matrix
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(conf_matrix, annot=True, fmt="d")
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation="sigmoid")
])
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(X_train, y_train, epochs=5)
```

```python
model.evaluate(X_test, y_test)
```

# Can you improve further the model's performance? Explain how.

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    
    keras.layers.Dense(512, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.3),
    
    keras.layers.Dense(256, activation="relu"),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.3),
    
    keras.layers.Dense(10, activation="softmax")
])
```

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)
```

```python
model.fit(X_train, y_train, epochs=30)
```

```python
model.evaluate(X_test, y_test)
```

```python
coef = model.get_weights()
```

```python
coef
```

In this model, I chose to use a deep feedforward architecture with two Dense layers (512 and 256 units) activated by ReLU to enable strong feature learning. I added BatchNormalization after each Dense layer to stabilize and speed up training, and included Dropout layers with a 30% rate to reduce overfitting. The final Dense layer uses a Softmax activation to output probabilities across the 10 MNIST classes. I also increased the number of epochs to 20 to give the model enough time to converge properly and reach a higher accuracy.
